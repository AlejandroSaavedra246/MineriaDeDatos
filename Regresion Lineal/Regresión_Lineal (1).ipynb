**REGRESION LINEAL**

Se realizo una regresion lineal en base a un documento de excel ya creado que contiene los datos de PSS y la edad como informacion para al final visualizar como se mostraria la regresion mediante una grafica gracias a la libreria "matplotlib.pyplot"

```python
#importamos las librerias
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#hacemos una conexion con drive
from google.colab import drive
drive.mount('/gdrive')

ruta = "/gdrive/MyDrive/Regresion Lineal/RegresiónLineal.csv"
data = pd.read_csv(ruta,encoding ="latin1")
```

Para crear la regresion Lineal vamos a trabajar con la edad y PSS sacado del excel que se nos compartio

```python
#Utilizamos dos variables que representarian a edad y PSS
X = data['Edad'].values.reshape(-1, 1)
Y = data['PSS'].values
```

```python
#Creamos un regresion lineal gracias a la libreria Linearregression
regresion_lineal = LinearRegression()
```

```python
#Mandamos a traer los valores de X and Y para la regresion
regresion_lineal.fit(X, Y)
```

```python
coeficientes = regresion_lineal.coef_
intercepcion = regresion_lineal.intercept_
```

```python
#aqui se a predecir para poder entrenar al modelo
predicciones = regresion_lineal.predict(X)
```

```python
#Generamos una grafica para mostrar la regresion lineal
plt.scatter(X, Y, color='red', label='Datos los Reales')
plt.plot(X, predicciones, color='Purple', label='Regresión Lineal')
plt.xlabel('Edad')
plt.ylabel('PSS')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```

CONCLUSION
Utilizando el excel de regreseion lineal la cual contaba con datos de Edad y PSS logramos hacer una regresion lineal donde se ven algunos datos dispersos y otros cerca de donde seria la regresion.
sklearn.linear_model utilizamos esta libreria para que el modelo conosca sobre regresion lineal
